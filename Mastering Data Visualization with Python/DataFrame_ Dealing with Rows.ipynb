# Single Condition      df[ condition.......]

```python
df[df['Length'] > 105]
```

# Multiple Condition filter  df[(cond1.......), (cond2.......)]

```python
df[(df['Length'] > 105) & (df['Operator'] == 'Op-1')]
```

# Finding the items which are between 95 and 105

```python
df[ (df['Length'] >= 95) & (df['Length'] <= 105) ]
```

# OR Condition       syntax df[ (), ()]

```python
df[ (df['Length'] >= 110) | (df['Width'] <= 45) ]
```
