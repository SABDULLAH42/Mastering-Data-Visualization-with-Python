```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pydataset import data
import seaborn as sns
```

```python
housing_df =  data('Housing')
```

```python
housing_df.head()
```

```python
sns.catplot(data = housing_df, x='bedrooms', y='price' , kind='bar');
```

```python
sns.catplot(data = housing_df, x='bedrooms', y='price' , kind='bar', hue='driveway',errorbar=('ci', False));
```

```python
sns.catplot(data = housing_df, x='bedrooms', kind='count');
```

```python
sns.catplot(data = housing_df, x='bedrooms', kind='count', col='driveway');
```

```python
sns.catplot(data = housing_df, x='bedrooms', y='price' , kind='point', hue='driveway',errorbar=('ci', False));
```

```python
sns.catplot(data = housing_df, x='bedrooms', y='price' , kind='point', hue='driveway', errorbar=('ci', 95));
```
