```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pydataset import data
```

```python
austres_df = data('austres')
austres_df.head()
```

```python
plt.plot(austres_df['time'], austres_df['austres'] );
```

```python
austres_df.head()
```

```python
##austres_df['austres'].pct_change()
```

```python
austres_df['pct_change'] = austres_df['austres'].pct_change()
```

```python
austres_df.head()
```

```python
plt.plot(austres_df['time'], austres_df['austres'] );
plt.plot(austres_df['time'], austres_df['pct_change'] );
```

# Subplot

```python
plt.subplots()
```

```python
fig, ax = plt.subplots()

fig.set_size_inches(10, 5)

ax.plot(austres_df['time'], austres_df['austres']);
ax.set_title('Australian Population')
ax.set_xlabel('Years')
ax.set_ylabel('Population (in thousands)');

fig.savefig('Australian Population.png', dpi=200)
```
