```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pydataset import data
```

```python
housing_df =  data('Housing')
```

```python
housing_df.head()
```

```python
fig, ax = plt.subplots()
ax.scatter(housing_df['lotsize'],  housing_df['price'])
ax.set_title('Lot Size vs. Price')
ax.set_xlabel('Lot Size')
ax.set_ylabel('Price');
```

```python
fig, ax = plt.subplots()
ax.hist(housing_df['price'], bins = 10)
ax.set_title('Histogram of house prices')
ax.set_xlabel('House Price')
ax.set_ylabel('Number of houses');
```

# Now creating the histogram and scatterplot together
histogram at the top
Scatterplot at the bottom

```python
fig, (ax1, ax2) = plt.subplots(2, 1)
ax1.hist(housing_df['price'], bins = 10)
ax1.set_title('Histogram of house prices')
ax1.set_xlabel('House Price')
ax1.set_ylabel('Number of houses');

ax2.scatter(housing_df['lotsize'],  housing_df['price'])
ax2.set_title('Lot Size vs. Price')
ax2.set_xlabel('Lot Size')
ax2.set_ylabel('Price');

fig.tight_layout()         ## to avoid overlapping
```

# Creating 4 plots here ie on a figure I will have 4 axis that will consists of histogram and scatterplot

```python
fig, ax = plt.subplots(2,2,    figsize=(10, 10 ) )

                                                          ## remember index = row
ax[0,0].hist(housing_df['price'],  bins = 10)            ## ax[0,0] mean index o, column 0 histogram is plotted
ax[0,0].set_ylabel ('Price')

ax[0,1].scatter(housing_df['lotsize'],  housing_df['price'] )  ## ax[0,1] index 0, col 1  scatter is plotted

ax[1,0].scatter(housing_df['lotsize'],  housing_df['price'] )    ## ax[1,0] index 1, col 0  scatter is plotted
ax[1,0].set_xlabel('Price')
ax[1,0].set_ylabel('Lot Size')

ax[1,1].hist(housing_df['lotsize'], bins=10 )              ## ax[1,1] mean index 1, column 1 histogram is plotted
ax[1,1].set_xlabel ('Lot Size')

fig.suptitle('Lot Size and Price Correlogram')
fig.tight_layout();
```
