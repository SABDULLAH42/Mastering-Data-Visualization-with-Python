```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pydataset import data
import seaborn as sns
```

```python
housing_df =  data('Housing')
```

```python
housing_df.head()
```

# Histogram

```python
sns.displot(data = housing_df, x='price');          ## kind='hist' by default in displot
```

# KDE plot

```python
sns.displot(data = housing_df, x='price', kind = 'kde'); 
```

# Now we want both the plot ie Histogram and KDE

```python
sns.displot(data = housing_df, x='price', kind='hist', kde=True); 
```

# All 3 in one plot ie Histogram, kde & rugplot

```python
sns.displot(data = housing_df, x='price', kind='hist', kde=True, rug=True); 
```

# ECDF: This basically give you the cumulative value

```python
sns.displot(data = housing_df, x='price', kind = 'ecdf'); 
```
