```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pydataset import data
import seaborn as sns
```

```python
housing_df =  data('Housing')
```

```python
housing_df.head()
```

```python
plt.figure(figsize = (5,5))
sns.histplot(data = housing_df, x= 'price');
```

```python
sns.displot(data = housing_df, x= 'price');
```

```python
sns.displot(data = housing_df, x= 'price', height = 4);
```

Aspect ratio is the ratio of height and width
Let say if aspect ratio is 2 , so that means this will become 2 times wider

```python
sns.displot(data = housing_df, x= 'price', height = 4, aspect=2);
```
