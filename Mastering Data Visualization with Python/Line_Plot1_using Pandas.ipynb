```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pydataset import data
```

```python
housing = data('Housing')
```

```python
housing
```

```python
housing.plot()
```

# Refining the Plot

```python
housing.plot(y='price');
```

```python
housing.plot(
    y='price',
    title='House Prices',
    xlabel='House Index',
    ylabel='Price',
    legend=False,
    figsize=(20,10)
);
```

```python
housing.plot(
    y=['price' , 'lotsize'],
    title='House Prices and Lot Size',
    xlabel='House Index',
    ylabel='Price',
    subplots=True
    
);
```

```python
housing.plot(
    y=['price' , 'lotsize'],
    title='House Prices and Lot Size',
    xlabel='House Index',
    ylabel='Price',
    subplots=True,
    layout=(1,2)
    
);
plt.tight_layout()
```
