```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pydataset import data
```

```python
austres_df = data('austres')
```

```python
austres_df.head()
```

# Plotting using Pandas

```python
austres_df.plot(x='time', y='austres');
    
```

# Plotting using Matplotlib

```python
plt.figure(figsize=(10,4))
plt.plot(austres_df['time'], austres_df['austres'])
plt.title('Australian resident population')
plt.xlabel('Year')
plt.ylabel('Population (in thousands)');
```

# using    .--b
b is the blue color,    -- double dotted line

```python
plt.figure(figsize=(10,4))
plt.plot(austres_df['time'], austres_df['austres'], '.--b');
```

```python
plt.figure(figsize=(10,4))
plt.plot(austres_df['time'], austres_df['austres'], '.--r');
```

# Using   .-r
r is the red color, - single continous line

```python
plt.figure(figsize=(10,4))
plt.plot(austres_df['time'], austres_df['austres'], '.-r');
```

# Using o-g
o is the marker, - is the line style, g is the color


```python
plt.figure(figsize=(10,4))
plt.plot(austres_df['time'], austres_df['austres'], 'o-g');
```
