```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pydataset import data
```

```python
housing = data('Housing')
```

```python
housing
```

```python
housing.groupby('bedrooms').count()
```

```python
bedroom_counts = housing.groupby('bedrooms')['bedrooms'].count()
```

```python
bedroom_counts
```

```python
bedroom_counts.plot(
    kind='bar',
    title='Bedroom Count',
    xlabel = 'Number of Bedrooms',
    ylabel = 'Count'
);
```
