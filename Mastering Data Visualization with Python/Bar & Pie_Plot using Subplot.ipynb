```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pydataset import data
```

```python
housing_df =  data('Housing')
```

```python
housing_df.head()
```

```python
story_counts = housing_df['stories'].value_counts()
```

```python
story_counts
```

# Creating Bar plot using subplot approach

```python
fig, ax = plt.subplots()
ax.bar(story_counts.index, story_counts);
ax.set_xticks(story_counts.index);
ax.set_title('Number of stories')
ax.set_xlabel('Stories')
ax.set_ylabel('Number of houses');
ax.grid( axis ='y', alpha= 0.5)
```

Creating Pie Chart side by side for the same thing
I need to make 2 plots ie 1 row and 2 columns

```python
fig, (ax1, ax2) = plt.subplots(1, 2,  figsize=(12, 6))
ax1.bar(story_counts.index, story_counts);
ax1.set_xticks(story_counts.index);
ax1.set_title('Number of stories')
ax1.set_xlabel('Stories')
ax1.set_ylabel('Number of houses');
ax1.grid( axis ='y', alpha= 0.5)


ax2.pie(story_counts, labels = story_counts.index, autopct = '%.2f%%');
ax2.legend()
ax2.set_title('Portion of houses with various stories')

fig.tight_layout()
```
