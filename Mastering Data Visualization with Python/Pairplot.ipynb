```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pydataset import data
import seaborn as sns
```

```python
iris_df = data('iris')
iris_df.head()
```

```python
sns.relplot(data = iris_df, x='Sepal.Length', y='Petal.Length', hue='Species');
```

```python
sns.displot(data = iris_df, x='Sepal.Length', hue='Species');
```

```python
sns.pairplot(data = iris_df);
```

```python
sns.pairplot(data = iris_df, hue='Species');
```

```python
sns.pairplot(data = iris_df, hue='Species', vars=['Sepal.Length', 'Petal.Length']);
```
