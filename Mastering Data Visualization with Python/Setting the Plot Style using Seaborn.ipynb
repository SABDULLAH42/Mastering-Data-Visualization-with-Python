```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pydataset import data
import seaborn as sns
```

```python
iris_df = data('iris')
iris_df.head()
```

```python
sns.set_style('dark')
sns.relplot(data = iris_df, x='Sepal.Length', y='Petal.Length', hue='Species');
```

```python
sns.displot(data = iris_df, x='Sepal.Length' );
```

```python
with sns.axes_style('whitegrid'):
    sns.displot(data = iris_df, x='Sepal.Length' );
```

```python
sns.set_style('white', {'ytick.left': True})
sns.relplot(data = iris_df, x='Sepal.Length', y='Petal.Length', hue='Species');
```
