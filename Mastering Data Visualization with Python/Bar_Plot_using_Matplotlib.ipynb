```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pydataset import data
```

```python
housing_df =  data('Housing')
```

```python
housing_df.head()
```

```python
housing_df.groupby('bedrooms')['bedrooms'].count()
```

```python
bedroom_counts = housing_df.groupby('bedrooms')['bedrooms'].count()
bedroom_counts
```

```python
plt.bar(
    x = bedroom_counts.index,
    height = bedroom_counts
);
plt.title('Houses by number of bedrooms')
plt.xlabel('Number of bedrooms')
plt.ylabel('Count')
plt.grid(axis = 'y', alpha = 0.5, linestyle='dotted')
```

# Pareto Chart

```python
bedroom_counts = housing_df.groupby('bedrooms')['bedrooms'].count()
bc_desc = bedroom_counts.sort_values(ascending = False)
bc_desc
```

```python
plt.bar(
    x = bc_desc.index ,
    height = bc_desc
);
plt.title('Houses by number of bedrooms')
plt.xlabel('Number of bedrooms')
plt.ylabel('Count')
plt.grid(axis = 'y', alpha = 0.5, linestyle='dotted')
```

# I expect the bar in descending order but didn't get We need to convert the index value into string using .astype(str)
because automatically sort the x value

```python
plt.bar(
    x = bc_desc.index.astype(str),
    height = bc_desc
);
plt.title('Houses by number of bedrooms')
plt.xlabel('Number of bedrooms')
plt.ylabel('Count')
plt.grid(axis = 'y', alpha = 0.5, linestyle='dotted')
```

```python
plt.bar(
    x = bc_desc.index.astype(str),
    height = bc_desc
);
plt.plot(bc_desc.index.astype(str), np.cumsum(bc_desc), 'o-r', alpha = 0.5)

plt.axhline(400)

plt.grid(axis = 'y', alpha = 0.5, linestyle='dotted')
```

# Now we are plotting bar and line chart
 in line plot x value = bc_desc.index.astype(str)
 y value =np.cumsum(bc_desc)       cumsum ie cumulative sum

```python
plt.bar(
    x = bc_desc.index.astype(str),
    height = bc_desc
);
plt.plot(bc_desc.index.astype(str), np.cumsum(bc_desc), 'o-r', alpha = 0.5)

plt.axhline(.80 * bc_desc.sum())

plt.grid(axis = 'y', alpha = 0.5, linestyle='dotted')
```

# 'o-r'  o is marker, - is linestyle and r is color
# bc_desc.sum() = total number of bedrooms

```python
plt.bar(
    x = bc_desc.index.astype(str),
    height = bc_desc
);
plt.plot(bc_desc.index.astype(str), np.cumsum(bc_desc), 'o-r', alpha = 0.5)

plt.axhline(.80 * bc_desc.sum(), alpha = 0.5, ls = ':', color ='red')

plt.grid(axis = 'y', alpha = 0.5, linestyle='dotted')
```

# Horizontal bar ie barh
# x axis will be replaced by y , and height will be replaced by width

```python
plt.barh(
    y = bedroom_counts.index,
    width = bedroom_counts
);
plt.title('Houses by number of bedrooms')
plt.ylabel('Number of bedrooms')
plt.xlabel('Count')
plt.grid(axis = 'x', alpha = 0.5, linestyle='dotted')
```
