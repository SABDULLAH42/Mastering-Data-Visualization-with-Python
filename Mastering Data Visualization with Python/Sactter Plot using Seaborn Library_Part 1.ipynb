```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pydataset import data
import seaborn as sns
```

```python
housing_df =  data('Housing')
```

```python
housing_df.head()
```

# using the relplot function of Seaborn Library

```python
sns.relplot(data = housing_df, x='lotsize', y='price', height=4 );
plt.title('House price vs lot-size')
plt.xlabel('Lot size')
plt.ylabel('Price')
plt.annotate('We can add text here', xy= (10_000, 150_000))
plt.axhline(125_000)
plt.grid(axis = 'y', alpha=0.5)
```
