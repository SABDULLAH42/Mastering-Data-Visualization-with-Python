```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pydataset import data
```

```python
austres_df = data('austres')
austres_df.head()
```

```python
austres_df['pct_change'] = austres_df['austres'].pct_change()
austres_df.head()
```

```python
fig, ax = plt.subplots()

fig.set_size_inches(10, 5)

ax.plot(austres_df['time'], austres_df['austres']);
ax.set_title('Australian Population')
ax.set_xlabel('Years')
ax.set_ylabel('Population (in thousands)');
```

```python
fig, (ax1, ax2) = plt.subplots(2,1, figsize =(10, 10))

ax1.plot(austres_df['time'], austres_df['austres']);
ax1.set_xlabel('Years')
ax1.set_ylabel('Population (in thousands)');


ax2.plot(austres_df['time'], austres_df['pct_change']);
ax2.set_xlabel('Years')
ax2.set_ylabel('Percent change');

fig.suptitle('Australian Polulation');
```

```python
fig, (ax1, ax2) = plt.subplots(2,1, figsize =(10, 10),  sharex=True)

ax1.plot(austres_df['time'], austres_df['austres'], alpha=0.5, ls=':', );
#ax1.set_xlabel('Years')
ax.set_ylabel('Population (in thousands)');


ax2.plot(austres_df['time'], austres_df['pct_change']);
#ax2.set_xlabel('Years')
ax2.set_ylabel('Percent change');
ax2.hlines(0.003, 1971.25, 1994)

fig.suptitle('Australian Polulation');
```

# Calulating the Mean value

```python
mean_pct_change = austres_df['pct_change'].mean()
mean_pct_change
```

```python
fig, (ax1, ax2) = plt.subplots(2,1, figsize =(10, 10),  sharex=True)

ax1.plot(austres_df['time'], austres_df['austres'], alpha=0.5, ls=':', );
#ax1.set_xlabel('Years')
ax.set_ylabel('Population (in thousands)');


ax2.plot(austres_df['time'], austres_df['pct_change']);
#ax2.set_xlabel('Years')
ax2.set_ylabel('Percent change');
ax2.hlines(mean_pct_change, austres_df['time'].min(), austres_df['time'].max() )
ax2.annotate('Mean % change', xy=(austres_df['time'].min(), mean_pct_change -0.0003) )

fig.suptitle('Australian Polulation');
```

```python
min_pct_change_y = austres_df['pct_change'].min()
min_pct_change_y
```

First we are filtering the pct_change column
Then we want the value which is equal to min_pct_change_y

```python
min_pct_change_x = austres_df[austres_df['pct_change'] == min_pct_change_y]
```

First we are filtering the pct_change column
Then we want the value which is equal to min_pct_change_y
then we want the value of time col only which we get in series
then use iloc method to get the first value from the index which was shown in Series earlier

```python
min_pct_change_x = austres_df[austres_df['pct_change'] == min_pct_change_y]['time'].iloc[0]
min_pct_change_x
```

```python
fig, (ax1, ax2) = plt.subplots(2,1, figsize =(10, 10),  sharex=True)

ax1.plot(austres_df['time'], austres_df['austres'], alpha=0.5, ls=':', );
#ax1.set_xlabel('Years')
ax1.set_ylabel('Population (in thousands)');


ax2.plot(austres_df['time'], austres_df['pct_change']);
ax2.set_xlabel('Years')
ax2.set_ylabel('Percent change');
ax2.hlines(mean_pct_change, austres_df['time'].min(), austres_df['time'].max() )
ax2.annotate('Mean % change', xy=(austres_df['time'].min(), mean_pct_change -0.0003) )

ax2.annotate('What happened here?',
            xy = (min_pct_change_x , min_pct_change_y),
            xytext = (min_pct_change_x -8, min_pct_change_y * 1.2 ),
            arrowprops = {}
            )
             

fig.suptitle('Australian Polulation');
```
