```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pydataset import data
import seaborn as sns
```

```python
iris_df = data('iris')
iris_df.head()
```

```python
sns.lmplot(data = iris_df, x='Sepal.Length', y='Petal.Length');
```

```python
sns.lmplot(data = iris_df, x='Sepal.Length', y='Petal.Length', hue='Species');
```

```python
sns.jointplot(data = iris_df, x='Sepal.Length', y='Petal.Length', kind='reg');
```
