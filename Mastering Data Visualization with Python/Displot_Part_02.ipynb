```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pydataset import data
import seaborn as sns
```

```python
housing_df =  data('Housing')
```

```python
housing_df.head()
```

```python
sns.displot(data = housing_df, x='price', kind = 'hist', hue='bedrooms'); 
```

```python
sns.displot(data = housing_df, x='price', kind = 'hist', hue='bedrooms', 
            multiple = 'stack', palette = 'colorblind'); 
```

```python
sns.displot(data = housing_df, x='price', kind = 'hist', hue='bedrooms', multiple = 'dodge'); 
```

```python
sns.displot(data = housing_df, x='price', kind = 'hist', hue='airco', multiple = 'dodge'); 
```

```python
sns.displot(data = housing_df, x='price', kind = 'kde', hue='bedrooms',
           palette = 'colorblind', aspect=2);
```
