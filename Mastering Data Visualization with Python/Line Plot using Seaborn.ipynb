```python
salaries_df = data('Salaries')
```

```python
salaries_df.head()
```

# Salaries of professor in the university plotting using Matplotlib

```python
plt.plot(salaries_df['yrs.service'], salaries_df['salary']);
```

# Salaries of professor in the university plotting using Seaborn

```python
sns.relplot(data =salaries_df, x='yrs.service', y='salary', kind='line');
```

```python
sns.relplot(data =salaries_df, x='yrs.service', y='salary', kind='line', aspect=2);
```

# We dont want to see the band or confidence interval(ci)

```python
sns.relplot(data =salaries_df, x='yrs.service', y='salary', kind='line', errorbar=('ci', False), aspect=2);
```

# Want to know there is a diff between the salary of male and female

```python
sns.relplot(data =salaries_df, x='yrs.service', y='salary', kind='line', 
            hue='sex', errorbar=('ci', False), aspect=2);
```

# Comparing the Salaries of Professors based on 2 Discipline

```python
sns.relplot(data =salaries_df, x='yrs.service', y='salary', kind='line', 
            hue='discipline', errorbar=('ci', False), aspect=2);
```

```python
sns.relplot(data =salaries_df, x='yrs.service', y='salary', kind='line', 
            hue='discipline', errorbar=('ci', False), aspect=2, col='sex');
```
