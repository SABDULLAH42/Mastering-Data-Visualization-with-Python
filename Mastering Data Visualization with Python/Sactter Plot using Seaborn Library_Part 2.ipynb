```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pydataset import data
import seaborn as sns
```

```python
housing_df =  data('Housing')
```

```python
housing_df.head()
```

```python
sns.relplot(data = housing_df, x='lotsize', y='price', height=4 );
plt.title('House price vs lot-size')
plt.xlabel('Lot size')
plt.ylabel('Price')
plt.annotate('We can add text here', xy= (10_000, 150_000))
plt.axhline(125_000)
plt.grid(axis = 'y', alpha=0.5)
```

# To see which houses are with driveway and which houses are without driveway usinh hue ie colour

```python
sns.relplot(data = housing_df, x='lotsize', y='price', kind='scatter', hue= 'driveway' );

```

# To see which houses are with driveway and which houses are without driveway using col ie column

col split the data separately

```python
sns.relplot(data = housing_df, x='lotsize', y='price', kind='scatter', col= 'driveway' );
```

# making the relationship between 4 variables by putting the row as well I will know whether the house has airconditioner
or doesn't have an airconditioner

```python
sns.relplot(data = housing_df, x='lotsize', y='price', kind='scatter', 
            col = 'driveway', row = 'airco', height= 3 );
```

```python
sns.relplot(data = housing_df, x='lotsize', y='price', kind='scatter', 
            col = 'driveway', row = 'airco', height= 3 , size='bedrooms');
```

# Size will differentiate the size of the dot based on number of bedrooms ie 1 bedroom with a smaller dot
and 6 bedroom with a bigger dot


## hue will differentiate the color based on stories

```python
sns.relplot(data = housing_df, x='lotsize', y='price', kind='scatter', 
            col = 'driveway', row = 'airco', height= 3 , size='bedrooms', hue='stories');
```
