```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pydataset import data
import seaborn as sns
```

```python
housing_df =  data('Housing')
```

```python
housing_df.head()
```

# Strip Plot

```python
sns.catplot(data = housing_df, x='bedrooms', y='price', hue='driveway');
```

Dodge will separate group of points with the driveway and without driveway

```python
sns.catplot(data = housing_df, x='bedrooms', y='price', hue='driveway', dodge=True);
```

# Swarm Plot

```python
sns.catplot(data = housing_df, x='bedrooms', y='price', kind='swarm', aspect=2.5);
```
