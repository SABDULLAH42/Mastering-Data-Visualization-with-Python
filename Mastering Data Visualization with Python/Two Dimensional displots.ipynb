```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pydataset import data
import seaborn as sns
```

```python
housing_df =  data('Housing')
```

```python
housing_df.head()
```

```python
sns.displot(data = housing_df, x='lotsize');
```

# Creating two dimensional histogram 

```python
sns.displot(data = housing_df, x='price', y='lotsize');
```

```python
sns.displot(data = housing_df, x='price', y='lotsize', cbar=True);
```

```python
sns.displot(data = housing_df, x='price', y='lotsize', kind='kde');
```
