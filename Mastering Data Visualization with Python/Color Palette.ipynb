```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pydataset import data
import seaborn as sns
```

```python
iris_df = data('iris')
iris_df.head()
```

```python
sns.set_style('white')
#options are: "darkgrid", "whitegrid", "dark", "white", "ticks"
sns.set_context('poster')
#options are: "paper", "notebook", "poster", "talk"
sns.relplot(data = iris_df, x='Sepal.Length', y='Petal.Length', hue='Species');
```

```python
housing_df = data('Housing')
```

```python
sns.relplot(data = housing_df, x='lotsize', y='price', hue='bedrooms');
```

```python
sns.relplot(data = housing_df, x='lotsize', y='price', 
            hue='bedrooms', palette='bright');
```
