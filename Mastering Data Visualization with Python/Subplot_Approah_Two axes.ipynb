```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pydataset import data
```

```python
austres_df = data('austres')
austres_df.head()
```

```python
plt.plot(austres_df['time'], austres_df['austres'] );
```

```python
austres_df['pct_change'] = austres_df['austres'].pct_change()
austres_df.head()
```

```python
plt.subplots()
```

```python
fig, ax = plt.subplots()

fig.set_size_inches(10, 5)

ax.plot(austres_df['time'], austres_df['austres']);
ax.set_title('Australian Population')
ax.set_xlabel('Years')
ax.set_ylabel('Population (in thousands)');
```

Now I want to make 2 plots one for the population and other for the percent change

2 axis are there ax1 and ax2 and kept in under tuples ()

plt.subplot(2,1) >> 2 row and 1 column 2 plots which are vertically aligned

plt.subplot(1,2) >> 1 row and 2 column for plotting side by side

fig.suptitle ie supertitle (title will be only 1 for the plot)

```python
fig, (ax1, ax2) = plt.subplots(2,1, figsize =(10, 10))

ax1.plot(austres_df['time'], austres_df['austres']);
ax1.set_xlabel('Years')
ax1.set_ylabel('Population (in thousands)');


ax2.plot(austres_df['time'], austres_df['pct_change']);
ax2.set_xlabel('Years')
ax2.set_ylabel('Percent change');

fig.suptitle('Australian Polulation');

```

Between the 2 axes xlabel value is common ie Year so we used sharex= True

```python
fig, (ax1, ax2) = plt.subplots(2,1, figsize =(10, 10),  sharex=True)

ax1.plot(austres_df['time'], austres_df['austres'], alpha=0.5, ls=':', );
#ax1.set_xlabel('Years')
ax1.set_ylabel('Population (in thousands)');


ax2.plot(austres_df['time'], austres_df['pct_change']);
#ax2.set_xlabel('Years')
ax2.set_ylabel('Percent change');

fig.suptitle('Australian Polulation');
```
